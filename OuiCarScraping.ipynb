```python
from bs4 import BeautifulSoup
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import requests
from random import uniform
import pandas as pd
from selenium.webdriver.chrome.options import Options
from selenium.webdriver import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver import ActionChains


comune_libelle=[]

df=pd.read_excel("RESSOURCE.xlsx")
i = 0
j = 0
print(df.shape)

for i in range(df.shape[0]):
    comune_libelle.append(str(df.iloc[i,0]))#df.iloc[i,2]
print(len(comune_libelle))
pages=[]
for i in range(df.shape[0]):
    if(comune_libelle[i]=="17000"):
        print(i)
        

for col in comune_libelle:  #
    DRIVER_PATH = "chromedriver.exe"
    driver = webdriver.Chrome(executable_path=DRIVER_PATH)
    driver.maximize_window()
    #driver.get("https://www.ouicar.fr/france?search_type=city&where="+col)
    driver.get("https://www.ouicar.fr/")
    time.sleep(6)
    driver.find_element_by_name("search-address").send_keys(col+", France")
    time.sleep(6)
    driver.find_element_by_id("didomi-notice-agree-button").click()
    time.sleep(6)
    driver.find_element_by_xpath("//button[@type='submit']").click()
    time.sleep(8)
    content = driver.page_source.encode('utf-8').strip()
    pages+= BeautifulSoup(content, "html.parser")
    driver.quit()
```

```python
from IPython.core.display import display, HTML
vehicule=[]
for i in range(len(pages)):
    print(comune_libelle[i])
    for j in range(len(pages[i].find_all('ul')[2].find_all("li"))):
        vehicule.append(pages[i].find_all('ul')[2].find_all("li")[j].find_all('figcaption')[0].find_all('div')[2].text+","+pages[i].find_all('ul')[2].find_all("li")[j].find_all('figcaption')[0].find_all('div')[3].text+","+pages[i].find_all('ul')[2].find_all("li")[j].find_all('figcaption')[0].find_all('div')[8].text)
```
